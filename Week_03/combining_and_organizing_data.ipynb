# Combining & organizing data

## Libraries and settings

```python
# Libraries
import os
import pandas as pd
import numpy as np
import fnmatch

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Show current working directory
print(os.getcwd())
```

## Importing the prepared rental apartments data

```python
# Read data to pandas data frame
df_orig = pd.read_csv('apartments_data_prepared.csv', 
                      sep=',', 
                      encoding='utf-8')

# Copy of data with selected colums
columns = [ 'web-scraper-order', 
            'address_raw',
            'rooms', 
            'area', 
            'luxurious',
            'price',
            'price_per_m2']
df = df_orig[columns]

# Get number of rows and columns
print(df.shape)

# Show first records
df.head(5)
```

## Combining data from different sources

### Reading rental apartment data with geocoded addresses

```python
# Meaning of variables
# lat: geographical latitude
# lon: geographical longitude
# bfs_number: official municipality id
# bfs_name: official municipality name

# Geocoded data (i.e. data with latitude and longitude)
df_geo = pd.read_csv('apartments_data_geocoded.csv', 
                     sep=',', 
                     encoding='utf-8')

# Rename columns
df_geo = df_geo.rename(columns={'web-scrape':'web-scraper-order',
                                'address_ra':'address_raw',
                                'BFS_NUMMER':'bfs_number', 
                                'NAME':'bfs_name'})

# Show data
df_geo[['web-scraper-order',
        'address_raw',
        'lat', 
        'lon', 
        'bfs_number', 
        'bfs_name']].head()
```

### Join geo-information to rental apartment data using .merge()

```python
df2 = df.merge(df_geo[['web-scraper-order', 
                       'lat', 
                       'lon', 
                       'bfs_number', 
                       'bfs_name']], 
               on="web-scraper-order")
df2.head()
```

### Reading municipality-level data

```python
# Meaning of variables:
# bfs_number: official municipality id
# bfs_name: official municipality name
# pop: number of residents (=population)
# pop_dens: population density (pop per km2)
# frg_pct: percentage foreigners
# emp: numer of employees

df_municip = pd.read_excel('municipality_data.xlsx', 
                           sheet_name='data_for_import')
df_municip.head(5)
```

### Join municipality data to rental apartment data using .merge()

```python
# Merge needs a key which must be identical in both data sets (here the key is 'bfs_number')
df3 = df2.merge(df_municip[['bfs_number', 
                            'pop', 
                            'pop_dens', 
                            'frg_pct', 
                            'emp',
                            'mean_taxable_income']],
                on="bfs_number")
df3.head(5)
```

### Export data to file

```python
# Remove missing values which may have ocured from the merging process
df3 = df3.dropna()

# Check for missing values per column
print(df3.isna().sum())

# Count number of rows and columns
print(df3.shape)

# Export apartment data to file
df3.to_csv('apartments_data_enriched.csv',
            sep=',',
            encoding='utf-8')
```

```python
# Sorting data by 'price' and 'area' with highest price above (ascending=False)
df3.sort_values(by=['price', 'area'], 
                ascending=False).head()
```

## Reshaping data

```python
df_sub = df3[['bfs_name', 'rooms', 'price', 'area']][:5]
print('Original shape')
print(df_sub, '\n')

df_sub_stacked = df_sub.stack()
print('Stacked')
print(df_sub_stacked, '\n')

# Using unstack
print('Unstacked (= back to original shape)')
print(df_sub_stacked.unstack())
```

### Pivoting data using .pivot_table()

```python
df3['price_per_m2'] = df3['price'] / df3['area']

# Using pivot_table to reshape the data and calculate means
pivot = pd.pivot_table(df3[['rooms', 'price', 'area', 'price_per_m2']],
               index=['rooms'],
               values=['price', 'area', 'rooms', 'price_per_m2'],
               aggfunc=np.mean)
```

### Compare the price with the price_per_m2

```python
pivot[['price', 'price_per_m2']]

# Interpretation
# The price_per_m2 decreases with increasing number of rooms. This is because larger apartments
# usually have a lower price per square meter than smaller apartments.
# For example, a 1-room apartment may have a price_per_m2 of 50 CHF/m2, while a 4-room apartment may have a price_per_m2 of only 29 CHF/m2.
```

### Create a new pivot table and use aggfunc='count' as the aggregation function

```python
pivot_count = pd.pivot_table(df3[['rooms', 'price']],
               index=['rooms'],
               values=['price'],
               aggfunc='count')

pivot_count.rename(columns={'price': 'count_apartments'}, inplace=True)
pivot_count

# Interpretation
# The number of apartments increases with the number of rooms. Reaching the peak at 3.5 rooms. and then decreases again.

# Plot for better visualization
import matplotlib.pyplot as plt

pivot_count.plot(
    kind='bar',
    figsize=(8, 4),
    legend=False,
    color='skyblue'
)

plt.title('Number of Apartments by Room Size')
plt.xlabel('Rooms')
plt.ylabel('Count of Apartments')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Create a new pivot table with the mean values of rooms, area, price, and price_per_m2 per municipality. Sort the pivot table by price and area with the most expensive municipalities on top.

```python
pivot_mean = pd.pivot_table(df3[['bfs_name', 'rooms', 'area', 'price', 'price_per_m2']],
               index=['bfs_name'],
               values=['rooms', 'area', 'price', 'price_per_m2'],
               aggfunc=np.mean)

pivot_mean = pivot_mean.sort_values(by=['price', 'area'], ascending=False)

pivot_mean
```
